# YOLOv5主函数解释

```python
# 打上了 --save-txt，--exist-ok etc..都表示这个被激活，为True，默认为False
# 可以print(opt)查看这些到底是True or False

if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    
    # --weights为权重，可以更换网络，例如yolov5s...
    parser.add_argument('--weights', nargs='+', type=str, default='yolov5l.pt', help='model.pt path(s)')
    
    # --source为所要处理的图片，打上文件夹的意思就是把文件夹里面的图片都进行处理，也可以打单个图片
    # 例如data/images/lemon/lemon1.jpg
    parser.add_argument('--source', type=str, default='data/images/lemon', help='source')  # file/folder, 0 for webcam
    
    # --img-size为设置图片初始大小，在yolov5的模型官网有推荐
    parser.add_argument('--img-size', type=int, default=640, help='inference size (pixels)')
    
    # --conf-thres为置信度，可以任意设置 0-1 之间的数值
    parser.add_argument('--conf-thres', type=float, default=0.25, help='object confidence threshold')
    
    # --iou-thres为交并比，设置同置信度
    parser.add_argument('--iou-thres', type=float, default=0.45, help='IOU threshold for NMS')
    
    # --device为GPU
    parser.add_argument('--device', default='', help='cuda device, i.e. 0 or 0,1,2,3 or cpu')
    
    # --view-img为在pycharm主界面显示处理之后的图像 “要在控制台打上：python detect.py --view-img” 再按回车就能运行
    parser.add_argument('--view-img', action='store_true', help='display results')
    
    # --save-txt为保存处理之后框的坐标位置，同要在控制台打上
    parser.add_argument('--save-txt', action='store_true', help='save results to *.txt')
    
    # --save-conf为保存处理之后的置信度，同要在控制台打上
    parser.add_argument('--save-conf', action='store_true', help='save confidences in --save-txt labels')
    
    # --nosave表示不存储处理后的图像信息
    parser.add_argument('--nosave', action='store_true', help='do not save images/videos')
    
    # --classes 指定图片只保留哪些对象 “--classes 0”表示只保存0类别的对象框
    parser.add_argument('--classes', nargs='+', type=int, help='filter by class: --class 0, or --class 0 2 3')
    
    # 增强nms的
    parser.add_argument('--agnostic-nms', action='store_true', help='class-agnostic NMS')
    
    # 增强检测的，实际效果不大
    parser.add_argument('--augment', action='store_true', help='augmented inference')
    
    # 没多大用
    parser.add_argument('--update', action='store_true', help='update all models')
    
    # --project 为改变结果存储的文件夹
    parser.add_argument('--project', default='runs/detect', help='save results to project/name')
    
    # --name 为改变结果存储的文件夹下面的小文件夹
    parser.add_argument('--name', default='exp', help='save results to project/name')
    
    # --exist-ok 如果为True，那么在上次处理后保存结果的文件夹下存储结果，如果为False，那么创建新的文件夹存储结果
    parser.add_argument('--exist-ok', action='store_true', help='existing project/name ok, do not increment')
    opt = parser.parse_args()
    print(opt)
    check_requirements(exclude=('pycocotools', 'thop'))

    with torch.no_grad():
        if opt.update:  # update all models (to fix SourceChangeWarning)
            for opt.weights in ['yolov5s.pt', 'yolov5m.pt', 'yolov5l.pt', 'yolov5x.pt']:
                detect()
                strip_optimizer(opt.weights)
        else:
            detect()
```
